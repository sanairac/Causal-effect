# Introduction

Causal effect can be defined as an outcome of a situation based on an occurence. The demonstration of causal effect in the real world can be linked to many everyday observations. For this particular project we will observe the causal effect of exposure to a virus, generating an epidemiological study by controlling certain variables. The setting for this study will be a gym, where viruses and infections are spread when gym-goers fail to wipe down their equipment after use. As several people use the same equipment over the course of the day, given one of them runs the probability of being infected by a virus, the virus will spread when healthy individuals use the same equipment used by infected individuals. A causal relationship will exist if the number of machines present in the gym directly impacts the number of people who fall ill with the virus.

The duration of the illness and the period of time over which the virus can survive on the equipment might also important variables, since they will directly impact the number of machines contaminated by the virus, which in turn impacts the number of people infected. Thus, the parameters of our study can be broken down into the following: $mv$ (machine virus contamination), $uv$ (user virus infection), $mt$ (machine contamination time) and $ut$ (user infection time). Each of of these four parameters can be further broken down into a $3D$ matrix, which accounts for time in days ($i$), the number of users or machines ($j$) and the presence of a virus ($k$). For example, $mv[i, j, k]$ will be considered $True$, when on day $i$, machine $j$ has virus $k$. A user immune to the virus ($ui$), on the other hand, will be accounted for as in a $2D$ array as $ui[i, j]$, which will be considered $True$ when user $i$ is immune to virus $j$. We will be assuming that the user becomes immune to the virus after they have been infected and recovered.

For our project we will be generating random numbers assigned to each machine user, given that each variable has a certain possibility of being true for infection, to distribute the probability of being infected and developing immunity across a series of trials for a given number of days. A causal relationship will emerge if we are able to conclude that one parameter effects the other. 

We will begin by importing the modules and packages that will define and implement the functions, classes, or variables that will be used over the course of this report:

```python
from numpy import *
import matplotlib.pyplot as plt
from tracerecorder import TraceRecorder
%matplotlib inline
```

```python
def step(itrial,t):
    global nusers,nmachines
    global mv,mt,uv,ut,ui
    global tr
    
    # create views on next day's array slices
    MV = mv[itrial,t+1,:,:]
    MT = mt[itrial,t+1,:,:]
    UV = uv[itrial,t+1,:,:]
    UT = ut[itrial,t+1,:,:]
    UI = ui[itrial,t+1,:,:]

    # copy current state to day t+1
    MV[:,:] = mv[itrial,t,:,:]
    MT[:,:] = mt[itrial,t,:,:]
    UV[:,:] = uv[itrial,t,:,:]
    UT[:,:] = ut[itrial,t,:,:]
    UI[:,:] = ui[itrial,t,:,:]
    
    # then modify state on day t+1
    MT[MV] += 1                             # age the germs on the infected machines by 1
    UT[UV] += 1                             # advance the infections in the users by 1
     
    MV[ MT >= viral_lifetime ]     = False  # let the old viruses on the machines die
    MT[ MT >= viral_lifetime ]     = 0    
                                            # let the long-infected users recover
    UV[ UT >= infection_duration ] = False
    UT[ UT >= infection_duration ] = 0  

    um = random.choice(range(nmachines),nusers,replace=True)

    for iu in range(nusers):
        user_viruses    = UV[iu].copy()       # save state to allow overwriting
        machine_viruses = MV[ um[iu] ].copy() # save state to allow overwriting
        MV[ um[iu] ][user_viruses]    = True  # contaminate machine with viruses on user
        MT[ um[iu] ][user_viruses]    = 0     # set age of contamination to 0
                                              # infect non-immune user with viruses on machine
       
        UV[ iu     ][machine_viruses  & logical_not(UI[iu])] = True 
        UT[ iu     ][machine_viruses  & logical_not(user_viruses)] = 0 # set infection age to 0
        UI[ iu, UV[iu] ] = True                                 # set user iu immune to future infection by current viruses

TRIAL_AXIS,DAY_AXIS,USER_AXIS,VIRUS_AXIS = 0,1,2,3
MACHINE_AXIS = 2
```

```python
# the following runs trials and makes a histogram of total number of users catching a virus over the whole time period

def bunch_of_trials():
    global uv,mv,ui,ut,mt
    global tr
    global nusers,nmachines
    days = range(ndays+1)
    imax = 0

    uv = zeros((ntrials,ndays+1,nusers   ,nspecies),dtype=bool)  # uv[l,i,j,k] = True if in trial l on day i user j infected with species k
    mv = zeros((ntrials,ndays+1,nmachines,nspecies),dtype=bool)  # mv[l,i,j,k] = True if in trial l on day i machine j contaminated with species k
    ui = zeros(uv.shape,dtype=bool) # on day i user j immune to species k in trial l 
    ut = zeros(uv.shape,dtype=int)  # days since infection
    mt = zeros(mv.shape,dtype=int)  # days since contamination


    tr = TraceRecorder(30,ndays)

    for it in range(ntrials):

        uv[it,0,0,0] = True  # initial infection

        for t in range(ndays):
            step(it,t)

        for iv in range(nspecies):
            tr.record(uv[it,:,:,iv].sum(axis=1) )
        imax = max(imax,uv.sum(axis=USER_AXIS).max())

    plt.figure(figsize=(11,5))
    plt.subplot(1,2,1)
    tr.imageplot(.5)
    plt.xlabel('day')
    plt.ylabel('number of users with virus')
    plt.title('darkess of dot indicates frequency of occurrence in trials');
    avg_infecteduserdays = uv.sum(axis=(TRIAL_AXIS,DAY_AXIS,USER_AXIS,VIRUS_AXIS))/ntrials
    print('Avg person-days of infection',avg_infecteduserdays)

    plt.subplot(1,2,2)
    ninfected = (uv.sum(axis=(DAY_AXIS,VIRUS_AXIS))>0).sum(axis=1)
    # histogram
    inffreqs = zeros(nusers+1)
    for n in ninfected:
        inffreqs[n] += 1
    plt.bar(range(nusers+1),inffreqs)
    plt.xlabel('number of users catching virus over entire time')
    plt.ylabel('frequency');
```

```python
def random_infect(ndays,nmachines,nusers,nspecies,prob):
    a = random.rand(ndays,nmachines,nusers,nspecies)
    uv = a<prob
    uv.sum(axis=1)
    didgetsick = uv.sum(axis=1)>0
    howmanygotsick = didgetsick.sum(axis=1)
    b = howmanygotsick.sum(axis=0)/(ndays*nusers)
    print(b)
```

The above function independently calculates the probability of a user being infected by the given virus. 

```python
nspecies = 1
nusers = 2
nmachines = 1              
ndays = 25
viral_lifetime = 3           
infection_duration = 2       
ntrials = 1000
vcolors = 'rgbmcy'
  
bunch_of_trials()
random_infect(ndays,nmachines,nusers,nspecies,0.02)  # the given probability of the virus infecting an individual will be 0.02.
```

We will begin our trials with the base assumption of two gym-goers using one machine, given one of them is infected with the virus, which takes two days to recover from and will remain active on the equipment it has come into contact with for the duration of three days. The average number of days a person was infected during this trial was four, since one piece of equipment was shared between the two users this outcome is not surprising- both users would be infected one after the other and would have become immune. The subplot on the right depicts that over the course of 1000 trials, both users were most certainly infected and the subplot on the left tells us that there was no variation in the manner in which this infection spread- the first user fell ill followed immidiately by the second, both recovered by the third day and there was no recorded occurence of the virus for the remaining $22$ days since both users had developed immunity. The probability of infection is $0.02$ during this particular trial run since there were only two users present, thus they both had equally narrow chance of infection.  

This is a very basic outline of how our trials shall be carried out. We will now increase the number of machines that are available to the two users:

```python
nusers = 2
nmachines = 5
  
bunch_of_trials()
random_infect(ndays,nmachines,nusers,nspecies,0.01)
```

As we increase the number of machines we can see an increase in variation in the graph on the left and the graph on the right tells us there is a greater chance for only one individual to fall ill rather than both. As a result the average number of days a user is infected rests at $2.97$, this is due to the fact that there is only one person out of the two falling ill for the majority of these trials. The probability of each user falling ill is still $0.02$ since there are only two users, however, we can observe that with an increase in the number of machines available there are fewer chances that the infected individual will come into contact with the healthy individual.

We will increase the number of machines available to the two users in the gym by a greater number, just to confirm that our observations are in fact correct:

```python
nusers = 2
nmachines = 15
 
bunch_of_trials()
random_infect(ndays,nmachines,nusers,nspecies,0.02)
```

Increasing the number of machines available to the two users to $15$ rapidly expanded the chances of only one person falling ill, as we can see from the subplot on the right. In comparison to the previous graph, the frequency of both users falling ill has fallen to below $300$ trials, whilst the frequency of only one person falling ill has increased to over $700$. We can see very little change in variation in the graph on the left, however, the average number of days for a user to fall ill has slightly decreased from $2.97$ to $2.402$. 

We have observed from the above trials that given there are $2$ users, the chance of one of them infecting the other is vastly reduced when there is a greater number of machines available to them in the gym. ## Conclusion

In conclusion, we have noticed that since there is correlation between the given variables in our trials, we can assume that a causal effect relationship is present between the number of users in the gym and the number of machines present. We have been able to show the possible influence between the two parameters of our trial by noting the impact a change in one variable directly impacted the other. It appears that with a larger number of machines than users present in a gym, a virus will have fewer chances of being spread, since not fewer users will have the probability of being exposed to it before infected users become immune and the virus dies out. Having a larger number of users in a gym than machines will mean that each user runs a greater probability of being exposed to the virus and thus being infected. However, in a gym with a large number of user and a small number of machines, all the users will be infected and develop immunity within a fewer number of trial days than will a gym that has roughly equal numbers of users and machines.
